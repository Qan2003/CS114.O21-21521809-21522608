## Xử lý cột 'judgement' thành 3 cột 'times', 'mems' và 'verdicts'

```python
import pandas as pd
import json

# Đọc file CSV
df = pd.read_csv('annonimized.csv')

# Chuyển đổi chuỗi JSON trong cột 'judgement' thành đối tượng Python
def safe_json_parse(item):
    try:
        json_obj = json.loads(item)
    except ValueError:
        return {}  # Trả về một đối tượng rỗng nếu JSON không hợp lệ
    else:
        # Kiểm tra xem các khóa cần thiết có tồn tại hay không
        if 'times' in json_obj and 'mems' in json_obj and 'verdicts' in json_obj:
            return json_obj
        else:
            return {}  # Trả về đối tượng rỗng nếu thiếu khóa

df['judgement'] = df['judgement'].apply(safe_json_parse)

# Tạo cột mới từ dữ liệu
df['times'] = df['judgement'].apply(lambda x: x.get('times') if x else None)
df['mems'] = df['judgement'].apply(lambda x: x.get('mems') if x else None)
df['verdicts'] = df['judgement'].apply(lambda x: x.get('verdicts') if x else None)

# Bỏ cột 'judgement' nếu bạn không cần nó nữa
df.drop(columns=['judgement'], inplace=True)

# Lưu vào file mới
df.to_csv('annonimized2.csv', index=False)
df
```

```python
import pandas as pd

# Đọc file CSV
df = pd.read_csv('annonimized2.csv')

# Hàm để chuyển chuỗi thành mảng số và tính tổng, trả về -1 nếu mảng rỗng
def sum_array(x):
    # Kiểm tra xem chuỗi có phải là mảng rỗng không
    if isinstance(x, str) and x.strip() == '[]':
        return -1
    # Xử lý chuỗi hoặc số, chấp nhận cả số nguyên và số thực
    try:
        if isinstance(x, str):
            # Chuyển chuỗi thành mảng số
            numbers = [float(num.strip()) for num in x.strip('[]').split(',')]
        else:
            # Nếu không phải chuỗi, giả định đó là một số đã sẵn sàng để tính toán
            numbers = [float(x)]
    except ValueError:
        return -1  # Trả về -1 nếu có lỗi trong quá trình chuyển đổi
    return sum(numbers)
# Áp dụng hàm này lên cột 'time'
df['times'] = df['times'].apply(sum_array)
df['mems'] = df['mems'].apply(sum_array)


# Lưu kết quả vào một file CSV mới
df.to_csv('annonimized2.csv', index=False)
```

## Lọc các username, nếu username có trong file public thì lưu vào train ngược lại lưu vào test.

```python
import pandas as pd

# Đọc dữ liệu từ file CSV 'qt-public.csv' và 'data.csv', lưu vào các biến tbtl_data và test_data.
tbtl_data = pd.read_csv('qt-public.csv')
test_data = pd.read_csv('data.csv')

# Tìm các tên người dùng chung giữa hai tập dữ liệu. Lọc ra những bản ghi từ test_data có tên người dùng xuất hiện trong tbtl_data.
common_usernames = test_data[test_data['username'].isin(tbtl_data['username'])]

# Tìm các tên người dùng chỉ có trong test_data mà không có trong tbtl_data bằng cách sử dụng phủ định của điều kiện isin.
remaining_usernames = test_data[~test_data['username'].isin(tbtl_data['username'])]

print(remaining_usernames['username'].nunique())
print(common_usernames['username'].nunique())

# Thực hiện ghép nối (merge) dữ liệu để thêm cột 'QT' từ tbtl_data vào common_usernames dựa trên cột 'username'.
# Chế độ 'left' đảm bảo rằng tất cả các bản ghi từ common_usernames sẽ được giữ lại, ngay cả khi không có dữ liệu tương ứng trong tbtl_data.
remaining_usernames_with_scores = pd.merge(common_usernames, tbtl_data[['username', 'QT']], on='username', how='left')

# Xuất dữ liệu của các tên người dùng chung ra file CSV mới tên là 'train_qt.csv' mà không lưu chỉ mục của DataFrame.
common_usernames.to_csv('train_qt.csv', index=False)

# Xuất dữ liệu của các tên người dùng không chung ra file CSV mới tên là 'test_qt.csv' mà không lưu chỉ mục của DataFrame.
remaining_usernames.to_csv('test_qt.csv', index=False)
```
